```python
#Diccionarios
planet = {
    'name' : 'Mars',
    'moons' : 2
}
```

```python
#Impresión del diccionario
print(planet['name'], "has", planet['moons'], "moons.")
```

```python
#Nuevos valores
planet['circumference (km)'] = {
    'polar' : 6752,
    'equatorial' : 6792
}
```

```python
#Impresión del planeta y su circunferencia polar
print(planet['name'], "has a polar circumference of",planet['circumference (km)']['polar'],"km.")
```

```python
# Planetas y lunas
planet_moons = {
    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1
}
```

```python
#Variables planets y moons usando values() y keys()
planets = planet_moons.keys()
moons = planet_moons.values()
```

```python
#Total de lunas y promedio
total_moons=0
for k in moons:
    total_moons += k
print(f"There are {total_moons} in the solar system,")
print(f"this means the average number of moons per planet is {round(total_moons/len(planets),2)}.")
```
