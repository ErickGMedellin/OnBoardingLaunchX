```python
#Función que lee tres tanques de combustible y muestra el promedio
def mean_tanks(tank1,tank2,tank3):
    mean1 = round((tank1+tank2+tank3)/3,2)
    return f"El promedio de los tanques es {mean1}"
```

```python
#Prueba de la función
print(mean_tanks(50,65,75))
```

```python
#Función promedio
def mean(list):
    con = 0
    for k in list:
        con += k
    return con/len(list)
```

```python
#Función actualizada
def mean_tanks_act(tank1,tank2,tank3):
    return f"El promedio de los tanques es {mean([tank1,tank2,tank3])}"
```

```python
#Prueba de la función actualizada
print(mean_tanks_act(50,60,70))
```

```python
#Definición de una función reporte de la misión
def mis_rep(pretime,flitime,destiny,extank,intank):
    return f"Destiny: {destiny} \nTotal travel time: {pretime+flitime} minutes \nTotal fuel: {extank+intank} gallons."
print(mis_rep(10,15,"Moon",2000,1200))
```

```python
#Nueva función de reporte considerando más argumentos
def mis_rep1(destiny,*min,**fuel1):
    return f"Destiny: {destiny} \nTotal travel time: {sum(min)} minutes \nTotal fuel: {sum(fuel1.values())} gallons."
print(mis_rep1("Moon",5,23,15,main=200,exter=500))
```

```python
#Nueva función que imprime también la información de cada tanque
def mis_rep2(destiny,*min,**fuel1):
    rep = f"Destiny: {destiny} \nTotal travel time: {sum(min)} minutes \nTotal fuel: {sum(fuel1.values())} gallons\n"
    for tank_name, gallons in fuel1.items():
        rep += f"{tank_name} tank --> {gallons} gallons left\n"
    return rep
print(mis_rep2("Mars",55,2300,705,main=20000,exter=585000))
```
