```python
text = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. 
On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. The highest daylight temperature of the Moon is 127 C."""
tex1=text.split(".")
```

```python
keyword = ["average","temperature", "distance"]
for i in keyword:
    for j in tex1:
        check = i in j
        if (check == True):
            print(j + ".\n")
```

```python
for j in tex1:
    check = j.find("C")
    if (check != -1):
        print(j)
        j = j.replace("C","Celsius")
        print(j)
```

```python
#Datos
name = "Moon"
gravity = 0.00162 #in km
planet = "Earth"
```

```python
#Título
title = f"Gravity facts about {name}"
```

```python
#Cadena larga de información de los hechos de la luna y la Tierra
facts = f"Planet name: {planet} \nGravity on {name}: {round(gravity*1000,2)} m/s2"
```

```python
#Unión de cadenas
print(f"{title} \n{facts}")
```

```python
#Cambio de los datos para probar la plantilla
#No se ejecuta como se desea por culpa del f
planet = "Marte"
gravity = 0.00143
name = "Ganímedes"
print(f"{title} \n{facts}")
```

```python
#Nueva plantilla con .format(variables)
grav = round(gravity*1000, 2)
new_template = "Gravity Facts about {nombre} \nPlanet Name : {planeta} \nGravity on {nombre}: {grave} m/s2"
print(new_template.format(nombre = name, planeta = planet, grave = grav))
```

```python
#Nueva plantilla con .format(variables)
#En este caso pondré 0, 1 y 2 para las variables
new_template = "Gravity Facts about {0} \nPlanet Name : {1} \nGravity on {0}: {2} m/s2"
print(new_template.format(name, planet, grav))
```
