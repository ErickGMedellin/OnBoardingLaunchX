```python
#Declaración de variables
planets = []
new_planet = ""
```

```python
#Ciclo while
while new_planet.lower() != "done":
    if new_planet:
        planets.append(new_planet)
    new_planet = input("Enter a planet's name or write done to finish")
```

```python
#Impresión de la lista con un ciclo for
for planet in planets:
    print(planet)
```
