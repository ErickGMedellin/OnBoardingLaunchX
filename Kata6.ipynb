```python
#Lista de los planetas y número de planetas en la lista
planets = ['Mercury','Venus','Earth','Mars','Jupiter','Saturn','Uranus','Neptune'] 
num_plan = len(planets)
print("There are", num_plan,"planets in the solar system")
```

```python
#Se agrega Plutón, se muestra el número de planetas y el último elemento
planets.append("Pluto")
print("There are actually", len(planets), "in the solar system.")
print("The last one is",planets[-1],".")
```

```python
#Se pide un planeta y se encuentra en la lista
inp_plan = input("Escribe el nombre de un planeta con la inicial mayúscula.")
plan_index = planets.index(inp_plan)
print(inp_plan, "is the planet number", plan_index+1,"nearest to the sun")
```

```python
#Se muestran los planetas más cercanos al sol antes del input
inp_plan = input("Escribe el nombre de un planeta con la inicial mayúscula.")
plan_index = planets.index(inp_plan)
print("These are the planets closer than", inp_plan, "to the sun:", planets[0:plan_index])
```

```python
#Se muestran los planetas más lejanos al sol después del input
inp_plan = input("Escribe el nombre de un planeta con la inicial mayúscula.")
plan_index = planets.index(inp_plan)
print("These are the planets further than", inp_plan, "to the sun:", planets[plan_index+1:])
```
